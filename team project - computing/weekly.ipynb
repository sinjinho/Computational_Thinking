```python
from tkinter import *

days = ["Ïõî", "Ìôî", "Ïàò", "Î™©", "Í∏à", "ÌÜ†", "Ïùº"]
todo_values = [[] for _ in range(7)]
todo_vars = []
todo_frames = []

root = Tk()
root.geometry("420x480")
main_frame = Frame(root)
default_frame = Frame(root)

def show_main_screen():
    default_frame.pack_forget()
    main_frame.pack(fill="both", expand=True)

def add_todo(day_idx):
    val = todo_vars[day_idx].get()
    if val.strip():
        todo_values[day_idx].append(val.strip())
        todo_vars[day_idx].set("")
        refresh_main_todo(day_idx)

def delete_todo(day_idx, value_idx):
    del todo_values[day_idx][value_idx]
    refresh_main_todo(day_idx)

def refresh_main_todo(day_idx):
    frame = todo_frames[day_idx]
    for widget in frame.winfo_children():
        widget.destroy()

    if todo_values[day_idx]:
        frame.pack(fill="x", padx=58, anchor="w")
        for idx, val in enumerate(todo_values[day_idx]):
            row = Frame(frame)
            row.pack(anchor="w", fill="x")
            Label(row, text=val, width=20, anchor="w").pack(side=LEFT)
            Button(row, text="üóë", command=lambda i=day_idx, j=idx: delete_todo(i, j)).pack(side=LEFT)
    else:
        frame.pack_forget()

def build_main_screen():
    Label(main_frame, text="ÏöîÏùºÎ≥Ñ Ìï† Ïùº Î™©Î°ù", font=("Arial", 16)).pack(pady=10)
    for i, day in enumerate(days):
        day_frame = Frame(main_frame)
        day_frame.pack(pady=6, fill="x", anchor="w")

        row = Frame(day_frame)
        row.pack(fill="x")
        Label(row, text=f"{day}ÏöîÏùº:", width=6, anchor="w").pack(side=LEFT)

        var = StringVar()
        entry = Entry(row, textvariable=var, width=18)
        entry.pack(side=LEFT)
        todo_vars.append(var)

        Button(row, text="+", command=lambda idx=i: add_todo(idx)).pack(side=LEFT, padx=3)

        val_frame = Frame(day_frame)
        todo_frames.append(val_frame)

build_main_screen()
show_main_screen()

root.mainloop()
```
